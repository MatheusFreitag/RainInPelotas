## Reading and Writing Data in Text Format

```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv('https://raw.githubusercontent.com/MatheusFreitag/RainInPelotas/master/Chuvas%20em%20Pelotas.csv')
```

```
df.columns=['Date', 'Status', 'F', 'C', 'url', 'icon']

df['C']= df['C'].map(lambda x: x[:2])
```

```
df['C']=df['C'].astype(float)
#Get a subset of the dataFrame
df = df[:20]

#Get temperature labels
scatter_labels = [temperature for temperature in df['C']]

#Generate Figure
ax = df.plot(x='Date', y='C', figsize=(16,7), colormap="autumn")

#Label of the y-Axis
ax.set_ylabel("Temperature in Celcius") 

#Set the min and max values in the y-Axis
ax.set_ylim(ymin=0, ymax=40)

#Which xtick should go as labels
xticklabels = [df['Date'][0], df['Date'][5], df['Date'][10], df['Date'][15],df['Date'][18]]

#Set the xticks but only at the right positions. The _ is to silent the output
_= ax.set_xticks([0,5,10,15,18]) 

#Set the labels of the xticks. The _ is to silent the output
_=ax.set_xticklabels(xticklabels, rotation=45) 


#Set the legend of the line
ax.legend(["Temperature in ÂºC"]); 

#Add label to points
for label, x, y in zip(scatter_labels, range(20), df['C'].astype(float)):
    plt.annotate(
        label,
        xy=(x, y), xytext=(0, -50),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.2),
        arrowprops=dict(arrowstyle = '-', connectionstyle='arc3,rad=0'))
```

```
df = pd.read_csv('https://raw.githubusercontent.com/MatheusFreitag/RainInPelotas/master/Chuvas%20em%20Pelotas.csv')

df.columns=['date', 'status', 'f', 'c', 'url', 'icon']

scatter_labels = (df['date'][df.index % 10 == 0]).map(lambda x: x[:-11])
```

```
#Treat Celsius column

df['c']= df['c'].map(lambda x: x[:2])
```

```
#Make temperature in Celcius a float type 
df['c']=df['c'].astype(float)

#Generate Figure
ax = df.plot(kind='bar', x='date', y='c', figsize=(20,7), colormap="autumn")

#Label of the y-Axis
ax.set_ylabel("Temperature in Celcius") 

#Set the min and max values in the y-Axis
ax.set_ylim(ymin=0, ymax=40)

#Which xtick should go as labels
xticklabels = scatter_labels.values

#Set the xticks but only at the right positions. The _ is to silent the output
_= ax.set_xticks(scatter_labels.index) 

#Set the labels of the xticks. The _ is to silent the output
_=ax.set_xticklabels(xticklabels, rotation=90) 


#Set the legend of the line
ax.legend(["Temperature in ÂºC"]); 
```
